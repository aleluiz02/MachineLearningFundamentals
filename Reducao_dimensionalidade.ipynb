```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL da imagem
url = 'https://wallpapers.com/images/high/super-amoled-display-b8phc12mkiq7yrok.webp'  # Substitua pela URL desejada

# Baixar a imagem da internet
response = requests.get(url)
img_original = Image.open(BytesIO(response.content)).convert('RGB')

# Converter para níveis de cinza (0 a 255)
img_gray = img_original.convert('L')

# Converter para binarizada (0 ou 255) usando um limiar
limiar = 10  # você pode ajustar esse valor
img_bw = img_gray.point(lambda x: 255 if x > limiar else 0, '1')

# Mostrar as imagens
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img_original)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Níveis de Cinza")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_bw, cmap='gray')
plt.title("Preto e Branco")
plt.axis('off')

plt.tight_layout()
plt.show()
```
